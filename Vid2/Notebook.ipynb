```python
df.groupby("City")["Average Cost for two"].mean().sort_values(ascending=False).head().plot(kind='bar', color='skyblue')
plt.title("Average Cost for Two by City")
plt.xlabel("City")
plt.ylabel("Amount")
plt.show()
```

```python
#Cuisines total no of votes
df.groupby("Cuisines")["Votes"].sum().sort_values(ascending=False).head(10).plot(kind='bar', color='lightgreen')
plt.title("Top 10 Cuisines by Total Votes")
```
